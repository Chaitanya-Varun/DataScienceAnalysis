```python
#Required Libraries
import numpy as np
import scipy.stats as ss
import matplotlib.pyplot as plt
```

# Problem-1

```python
#Given Parameters
# Measurement-1
Ho_1 = 67
Ho_1_err = 2.3
# Measurement_2
Ho_2 = 71
Ho_2_err = 1.3
# z-score
# z = (|x1 - x2|) / sqrt(s1^2 + s2^2)
z = abs(Ho_1 - Ho_2)/np.sqrt(Ho_1_err**2+ Ho_2_err**2)
print("Z-score:",z)
```

Therefore, the significance of the discrepancy between the measurements is approximately 1.52 sigmas. This means that the discrepancy is not statistically significant, as it is less than the commonly accepted threshold of 3 sigmas.

```python
# Given P
M =2
N = 10
dof = N-M
p_val = 0.15 #CI:85%
delta_chi_sq = ss.chi2(dof).ppf(p_val)
print("Delta Chi Square:",delta_chi_sq)
print("Verification by ss.chi2(dof).sf(delta_chi_sq) gives:",ss.chi2(dof).sf(delta_chi_sq))
```

# Problem-3

```python
def BIC(L, params, N):
    return -2*np.log(L) + params*np.log(N)
```

```python
# Model-1
chi_a= 1.3
chi_b = 0.7
dof_a = 6
dof_b = 8
N = 10 
bic_a = BIC(chi_a, dof_a, N)
bic_b = BIC(chi_b, dof_b, N)
print("BIC-A: ",bic_a)
print("BIC-B: ",bic_b)
del_bic = abs(bic_a-bic_b)
print("Delta BIC: ",del_bic)
```
